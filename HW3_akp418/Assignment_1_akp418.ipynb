# Normal distribution


```python
mean_normal = 100 #mu = 100
stddev = np.std(size)

x = np.random.normal(mean_normal, stdev, 2000)
count, bins, ignored = plt.hist(x, 26, normed=True)

plt.plot(10, 1/(stddev * np.sqrt(2 * np.pi)) * np.exp( - (10 - mean_normal)**2 / (2 * stddev**2) ), linewidth=2)
plt.xlabel('X')
plt.ylabel('N')
pl.title('Normal', fontsize = 18, fontweight = 'bold')
pl.text(-800,-0.0004, 'Figure 3:', fontsize = 15, fontweight = 'bold')
pl.text(-250,-0.0004, 'Normal Distribution : Histogram with mean at 100', fontsize = 14)
plt.show()

normal_dist = []
for i in range(0, len(size)):
    normal_dist.append(np.random.normal(mean_normal, stddev, size[i]))

normal_means = []
for i in range(0 , len(size)):
    normal_means.append(normal_dist[i].mean())
```

# Do it for the Poisson distribution

```python
mean_poisson = 100

s = np.random.poisson(mean_poisson, 2000)
count, bins, ignored = plt.hist(s, 26, normed=True)
plt.xlabel('X')
plt.ylabel('N')
pl.title('Poisson', fontsize = 18, fontweight = 'bold')
pl.text(70,-0.015, 'Figure 5:', fontsize = 15, fontweight = 'bold')
pl.text(90,-0.015, 'Poisson Distribution : Histogram with mean at 100', fontsize = 14)
plt.show()

poisson_dist = []
for i in range(0, len(size)):
    poisson_dist.append(np.random.poisson(mean_poisson, size[i]))

poisson_means = []
for i in range(0 , len(size)):
    poisson_means.append(poisson_dist[i].mean())
    
poisson_dist_plot = pl.figure(figsize = (10,6)).add_subplot(111)
poisson_dist_plot.plot(size, poisson_means, 'o')
poisson_dist_plot.set_xlabel('Sample Size', fontsize = 18)
poisson_dist_plot.set_ylabel('Sample Mean', fontsize = 18)
poisson_dist_plot.set_title('Poisson Distribution', fontsize = 18 , fontweight = 'bold')
poisson_dist_plot.text(0,92, 'Figure 6:', fontsize = 15, fontweight = 'bold')
poisson_dist_plot.text(300,92, 'Poisson Distribution : Scatter Plot with mean at 100', fontsize = 14)
poisson_dist_plot.plot([min(size), max(size)], [mean_poisson, mean_poisson])
```
